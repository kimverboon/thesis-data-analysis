```python
# Visual Analogue Scale (VAS)
# Pre-intervention / 360VR

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import shapiro

# read file
df = pd.read_csv('pre-intervention-values.csv', sep=';')

# extract columns containing participant IDs and VAS scores
vas_df = df[['Q1', 'Q1_1']].copy()
vas_df.columns = ['Participant_ID', 'VAS_Score']
vas_df = vas_df.drop_duplicates(subset='Participant_ID')

# convert to numeric data (to avoid errors)
vas_df['Participant_ID'] = pd.to_numeric(vas_df['Participant_ID'], errors='coerce')
vas_df['VAS_Score'] = pd.to_numeric(vas_df['VAS_Score'], errors='coerce')
vas_df = vas_df.dropna()

# filter only participants in the 360VR condition (so IDs 31 â€“ 60)
vas_df = vas_df[(vas_df['Participant_ID'] >= 31) & (vas_df['Participant_ID'] <= 60)]

# print(vas_df['VAS_Score'].describe())
vas_df = vas_df.groupby('Participant_ID')['VAS_Score'].mean().reset_index()

# now calculate the descriptive statistics
mean_vas = vas_df['VAS_Score'].mean()
std_vas = vas_df['VAS_Score'].std()
median_vas = vas_df['VAS_Score'].median()

# and print the specific descriptive statistics
print(f"Mean={mean_vas:.2f}")
print(f"SD={std_vas:.2f}")
print(f"Median={median_vas:.2f}")

# now shapiro wilk test for normality
stat, p_value = shapiro(vas_df['VAS_Score'])
print(f"Shapiro-Wilk test: T={stat:.2f}, p-value={p_value:.3f}")

# plot the histogram 
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
sns.histplot(vas_df['VAS_Score'], kde=True, bins=10, color='skyblue', edgecolor='black')
plt.axvline(mean_vas, color='red', linestyle='--', linewidth=2, label=f'Î¼ = {mean_vas:.2f}')
plt.legend(fontsize=12, loc='upper right')
plt.title('Pre-intervention VAS Scores \n(360VR)', fontsize=16)
plt.xlabel('Mean VAS Score', fontsize=12)
plt.ylabel('Num. of participants', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# comparison pre/post for 360VR or audio-only
from scipy.stats import ttest_rel

pre = pd.read_csv('pre-intervention-values.csv', sep=';')
post = pd.read_csv('post-intervention-values.csv', sep=';')

# prep the datasets by selecting again the Participant_ID and VAS score
pre_vas = pre[['Q1', 'Q1_1']].copy()
pre_vas.columns = ['Participant_ID', 'VAS_Score_Pre']
pre_vas = pre_vas.drop_duplicates(subset='Participant_ID')

post_vas = post[['participant_ID', 'Q297_1']].copy()
post_vas.columns = ['Participant_ID', 'VAS_Score_Post']
post_vas = post_vas.drop_duplicates(subset='Participant_ID')

# convert to numeric
pre_vas['Participant_ID'] = pd.to_numeric(pre_vas['Participant_ID'], errors='coerce')
pre_vas['VAS_Score_Pre'] = pd.to_numeric(pre_vas['VAS_Score_Pre'], errors='coerce')
pre_vas = pre_vas.dropna()

post_vas['Participant_ID'] = pd.to_numeric(post_vas['Participant_ID'], errors='coerce')
post_vas['VAS_Score_Post'] = pd.to_numeric(post_vas['VAS_Score_Post'], errors='coerce')
post_vas = post_vas.dropna()

# now merge pre and post 
merged = pd.merge(pre_vas, post_vas, on='Participant_ID')

# convert to numeric
merged['VAS_Score_Pre'] = pd.to_numeric(merged['VAS_Score_Pre'], errors='coerce')
merged['VAS_Score_Post'] = pd.to_numeric(merged['VAS_Score_Post'], errors='coerce')

# filter if you want 360VR or audio-only (in this case 360VR)
merged = merged[(merged['Participant_ID'] >= 61) & (merged['Participant_ID'] <= 90)]

# paired t-test
t_stat, p_value = ttest_rel(merged['VAS_Score_Pre'], merged['VAS_Score_Post'])
print(f"T={t_stat:.2f}, p={p_value:.3f}")

# Cohen's d for paired t-test
mean_diff = merged['VAS_Score_Pre'].mean() - merged['VAS_Score_Post'].mean()
std_diff = merged['VAS_Score_Pre'].std()
cohen_d = mean_diff / std_diff
print(f"Cohen's d ={cohen_d:.2f}")
```

```python
# now comparing 360VR vs audio-only

from scipy.stats import ttest_ind

# again load pre and post data 
pre = pd.read_csv('pre-intervention-values.csv', sep=';')
post = pd.read_csv('post-intervention-values.csv', sep=';')

# prep pre data
pre_vas = pre[['Q1', 'Q1_1']].copy()
pre_vas.columns = ['Participant_ID', 'VAS_Pre']
pre_vas['Participant_ID'] = pd.to_numeric(pre_vas['Participant_ID'], errors='coerce')
pre_vas['VAS_Pre'] = pd.to_numeric(pre_vas['VAS_Pre'], errors='coerce')
pre_vas = pre_vas.dropna()
merged['VAS_Change'] = merged['VAS_Score_Post'] - merged['VAS_Score_Pre']

# prep post data
post_vas = post[['participant_ID', 'Q297_1']].copy()
post_vas.columns = ['Participant_ID', 'VAS_Post']
post_vas['Participant_ID'] = pd.to_numeric(post_vas['Participant_ID'], errors='coerce')
post_vas['VAS_Post'] = pd.to_numeric(post_vas['VAS_Post'], errors='coerce')
post_vas = post_vas.dropna()

# merging pre and post
merged = pd.merge(pre_vas, post_vas, on='Participant_ID')
merged['VAS_Change'] = merged['VAS_Post'] - merged['VAS_Pre']

# split into the conditions
vr360 = merged[(merged['Participant_ID'] >= 31) & (merged['Participant_ID'] <= 60)]
audio = merged[(merged['Participant_ID'] >= 61) & (merged['Participant_ID'] <= 90)]

# independent t-test
t_stat, p_val = ttest_ind(vr360['VAS_Change'], audio['VAS_Change'], equal_var=False)  # Welchâ€™s t-test
print(f"T={t_stat:.2f}, p-value: {p_val:.3f}")
```
